#### String Formatting Methods

Methods
* capitalize() - capitalizes the first character of a string
* lower() - all characters of a string are made lowercase
* upper() - all characters of a string are made uppercase
* swapcase() - all characters of a string are made to switch case upper becomes lower and vice versa
* title() - each 'word' separated by a space is capitalized

```python
# In line .upper()
print( name1.upper())
print(name1)
```

```python
# In memory .lower()
name1 =  name1.lower()
print(name1)
```

```python
# del
del name1
```

#### String searching

```python
# .find with one and two Pakistan
p = "We are pakistan. We love our country Pakistan"
print(p.find("Pakistan"))
p = "We are Pakistan . We love our country Pakistan"
print(p.find("Pakistan"))
```

**String Indexing**

```python
# for Pakistan
p[7:15:1] #start:end:step        --> end not included
```

```python
# Step and Reversal
print(p[::2]) #start:end:step        --> end not included
p[::-1]
```

```python
# Smart Indexing
print(p[p.find("Pakistan"):15])
```
